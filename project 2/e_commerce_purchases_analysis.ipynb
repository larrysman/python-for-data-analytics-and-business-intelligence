**15. `How many people have a Credit Card that expires in 2020?`**

```python
data.columns
```

```python
data["CC Exp Date"]
```

###### **because of the expiry date type 02/20..02/19, we would need to split by `/` to access the year which is a second value in this dataset.**

```python
data["CC Exp Date"].str.split("/")
```

###### **Write a function to perform the splitting and extract the second item in the list `[02, 20]` using indexing, here `02` is indexed `0` and `20` is indexed `1`**

```python
def date_splitter(date_string):
    return date_string.split("/")[1]
```

```python
data["CC Exp Date"].apply(date_splitter)
```

```python
data["CC Exp Date"].apply(date_splitter) == "20"
```

```python
len(data[data["CC Exp Date"].apply(date_splitter) == "20"])

print(f"The number of people whose credit card expires in 2020 is: {len(data[data['CC Exp Date'].apply(date_splitter) == '20'])}.")
```

##### **USING THE ANONYMOUS LAMBDA FUNCTION**

```python
data["CC Exp Date"].apply(lambda date_string: date_string.split("/")[1] == "20")
```

```python
len(data[data["CC Exp Date"].apply(lambda date_string: date_string.split("/")[1] == "20")])
```

##### **A MORE COMPACTED FUNCTION AND PRINT THE RESULT AT ONCE**

```python
def date_splitter_and_counter():
    counter = 0
    for date_string in data["CC Exp Date"]:
        if date_string.split("/")[1] == "20":
            counter = counter + 1
    
    print("The number of people whose credit card expires in 2020 is: ", counter)
```

```python
date_splitter_and_counter()
```

**16. `The top 5 most popular email providers with domain name such as gmail.com, yahoo.com, etc`**

```python
data.columns
```

```python
data["Email"]
```

###### **We would need to extract the email domian name from each email by splitting and accessing the second value**

```python
data['Email'].str.split('@')
```

```python
list_of_email_providers = []

for domain_name in data['Email']:
    split = domain_name.split('@')[1]
    list_of_email_providers.append(split)
```

```python
list_of_email_providers[:4]
```

###### **Create a temporary column for the email providers in the data**

```python
data['Email Providers'] = list_of_email_providers

data.head(2)
```

###### **Then use value_counts() to diplay the top 5**

```python
data.columns
```

```python
top_5_email_providers = data['Email Providers'].value_counts().head()
```

```python
print("The top 5 Email Providers are: \n", top_5_email_providers)
```
