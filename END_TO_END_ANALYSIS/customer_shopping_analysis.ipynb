###### *There are missing values in the dataset for the `review rating column` howbeit, this missing values can be computed using `mean or median` but it is worth to know that the `mean` is proned to outliers hence may not be the best metric to use for correcting the missing values instead the `median` would be the best option here but here is another point to watch out for, this rating is for different categories hence using `median` or `mean` to as a single replacement across the entire dataset may introduce biases, hence the best approach is to compute the `median` for each categories and use it to replace it.*

```python
df.groupby('Category')['Review Rating'].transform(lambda category: category.fillna(category.median()))
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda category: category.fillna(category.median()))
```

###### **`Create another column with the age called the age_group - ['Young Adult', 'Adult', 'Middle Aged', 'Senior']`**

```python
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

###### **`Create another column for purchase_frequency_days using frequency_of_purchases`**

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
df['freq_of_purchases_in_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'freq_of_purchases_in_days']].head(10)
```
