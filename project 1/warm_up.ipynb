**13. `Apply Method`**

*In using apply method in Python, you just need to write a function that perform specific task and apply the function to the dataframe*

**Write and apply a function to half the scores of students**

```python
def half_score(score):
    return score / 2
```

```python
data["Scores"].apply(half_score)
```

```python
data["Half_Score"] = data["Scores"].apply(half_score)
data
```

**Write and apply function that add a bonus mark of 10 to any students whose half score is below 40**

```python
def bonus(scores):
    if scores < 40:
        return scores + 10
    else:
        return scores
```

```python
data["Half_Score"].apply(bonus)
```

```python
data["bonus_Score"] = data["Half_Score"].apply(bonus)
data
```

**USING APPLY METHOD AND THE ANONYMOUS FUNCTION CALLED LAMBDA**

```python
lambda_half_score_function = lambda score: score // 2
data["Scores"].apply(lambda_half_score_function)
```
