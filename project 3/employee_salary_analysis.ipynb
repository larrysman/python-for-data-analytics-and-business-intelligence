###### *We expect the following columns to be a float or int data types `['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']` but they were all strings and we need to convert them back to float or int datatypes*

###### *CONVERT NUMERICAL COLUMNS IN DTYPE (STR) TO DTYPE (FLOAT64)*

**10. `The total number of Job Titles that contain Captain`**

```python
data.columns
```

```python
def extract_captain(strings):
    if 'captain' in strings.lower():
        return strings
    else:
        pass
```

```python
data['JobTitle'].apply(extract_captain)
```

```python
# DROP THE NaN

data['JobTitle'].apply(extract_captain).dropna()
```

```python
# NOW COUNT THE NUMBER OF JOB TITTLE THAT CONTAIN CAPTAIN IN THE DATASET USING THE len() FUNCTION
len(data['JobTitle'].apply(extract_captain).dropna())
```

```python
print(f"The number of Job Title with the word 'captain' in the job title is {len(data['JobTitle'].apply(extract_captain).dropna())}.")
```

```python
# USING LAMBDA FUNCTION
len(data['JobTitle'].apply(lambda jobtitle: jobtitle if 'captain' in jobtitle.lower() else None ).dropna())
```

```python
# PASSED UNTO THE DATAFRAME TO EXTRACT ONLY THE TRUE VALUES

data[data['JobTitle'].str.contains('captain', case=False, na=False)]
```
